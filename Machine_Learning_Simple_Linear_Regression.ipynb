### Columns information:
- Year : Mentioned Year

-  WinterRain  : Winter Rainfall (in mm)

- HarvestRain  : Harvest Temprature

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Nikhil07-source/Machine-Learning/refs/heads/main/wine.csv')
df
```

```python
X = df[['AGST']]  #independent feature
y = df[['Price']] #dependent feature, target
```

##Splitting the data in 2 tables:
- Train
- Test

Usually it is done in the ration of 70:30 or 80:20.

```python
from sklearn.model_selection import train_test_split  #using sklearn library to split dataframe in train and test.

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=123)

X_test

# In the above code our dataframe is getting splitted in 2 tables - train,test.
# We are splitting dataframe in 80:20 ratio.
# Random State is used to get result same everytime, if we will not use this we will get different result everytime.
```

```python
#Checking the shape of train & test data
print('The Shape of X_train data is: ', X_train.shape)
print('The Shape of X_test data is: ', X_test.shape)
print('The Shape of y_train data is: ', y_train.shape)
print('The Shape of y_test data is: ', y_test.shape)
```

```python
#Training the data using Sklearn library
from sklearn.linear_model import LinearRegression

lm = LinearRegression() #model object

#model training/ model fitting
lm.fit(X_train, y_train)
```

###Below this the code is just for understanding, we don't need to do this much calculation manually.

```python
lm.coef_
```

```python
lm.intercept_
```

```python
#y = bx + c #here b represent Beta
#Price = 0.74 * AGST -5.13
```

###Now we will do prediction on train data

```python
#Doing predection on train data
prediction_train = lm.predict(X_train)

prediction_train
```

### Model Evaluation - (02:27:00)

```python
from sklearn import metrics
from math import sqrt
```

```python
# Actual value = y_train
# Prediction   = prediction_train

mse_train = metrics.mean_squared_error(y_train, prediction_train)
print('Mean Squared Error: ', round(mse_train, 3))                          #round(var, digit) = It will do roundoff till 3 digits.

rmse_train = sqrt(mse_train)
print('Root Mean Square Error: ', round(rmse_train, 3))                     #round(var, digit) = It will do roundoff till 3 digits.

#We can also use below code to caluclate rmse_train
#rmse_train = metrics.root_mean_squared_error(y_train, prediction_train)

mae_train = metrics.mean_absolute_error(y_train, prediction_train)
print('Mean Absolute Error: ', round(mae_train, 3))                         #round(var, digit) = It will do roundoff till 3 digits.

mape_train = metrics.mean_absolute_percentage_error(y_train, prediction_train)
print('Mean Absolute Percentage Error: ', round(mape_train, 3))             #round(var, digit) = It will do roundoff till 3 digits.

R2_train = metrics.r2_score(y_train, prediction_train)
print('R2 score: ', round(R2_train, 3))
```
